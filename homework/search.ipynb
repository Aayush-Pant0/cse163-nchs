# Search

In this homework, you'll implement a basic search engine by defining your own Python classes. A **search engine** is an algorithm that takes a query and retrieves the most relevant documents for that query. In order to identify the most relevant documents, our search engine will use **term frequency–inverse document frequency** ([tf–idf](https://en.wikipedia.org/wiki/Tf%E2%80%93idf)), an information statistic for determining the relevance of a term to each document from a corpus consisting of many documents.

The **tf–idf statistic** is a product of two values: term frequency and inverse document frequency. **Term frequency** computes the number of times that a term appears in a **document** (such as a single Wikipedia page). If we were to use only the term frequency in determining the relevance of a term to each document, then our search result might not be helpful since most documents contain many common words such as "the" or "a". In order to downweight these common terms, the **document frequency** computes the number of times that a term appears across the **corpus** of all documents.

````python
%%ipytest -v

class Document:
    """
    Creates documents which store a documents information regarding the frequency of each word within it
    """

    def __init__(self, path: str) -> None:

        self._path: str = path
        self._frequencies: dict[str, float] = {} # Actual word frequencies
        self._word_counts: dict[str, int] = {} # Number of times each word occurs in document

        with open(path) as f:
            terms = [clean(word) for word in f.read().split() if clean(word) != ""]

        if len(terms) > 0:

            for term in terms:
                self._word_counts[term] = self._word_counts.get(term, 0) + 1 # Creating counts for each word in document

            for term in terms:
                self._frequencies[term] = self._word_counts[term] / len(terms) # Count / Total = Frequency
    
    def term_frequency(self, term: str) -> float:
        """
        Takes a term and returns the frequency of that term in the specified document
        """

        # Return frequency if there, otherwise 0
        if clean(term) in self._frequencies.keys():
            return self._frequencies[clean(term)]

        return 0.0
    
    def get_path(self) -> str:
        """
        Gives the path of the file that the document object is based on 
        """

        return self._path
    
    def get_words(self) -> set[str]:
        """
        Gives a list of the unique words within the document specified as a set
        """

        return set(self._frequencies)

    def __repr__(self) -> str:
        """
        Gives a string format for representing the document
        """

        return f'Document(\'{self._path}\')'


class TestDocument:
    doc1 = Document("doggos/doc1.txt")
    euro = Document("small_wiki/Euro - Wikipedia.html")
    test1 = Document("test_files/test_1.txt")
    test2 = Document("test_files/test_2.txt")
    test3 = Document("test_files/test_3.txt")

    def test_term_frequency(self) -> None:
        assert self.doc1.term_frequency("dogs") == pytest.approx(1 / 5)
        assert self.euro.term_frequency("Euro") == pytest.approx(0.008651740942156932)
        assert self.test1.term_frequency("bi!!@#@#!g") == pytest.approx(3/10)
        assert self.test2.term_frequency("chicke!n##s") == pytest.approx(3/10)
        assert self.test3.term_frequency("gre@at") == pytest.approx(2/10)        

    def test_get_words(self) -> None:
        assert self.doc1.get_words() == set("dogs are the greatest pets".split())
        assert set(w for w in self.euro.get_words() if len(w) == 1) == set([
            *"0123456789acefghijklmnopqrstuvxyz".lower() # All one-letter words in Euro
        ])
        assert self.test1.get_words() == set("the other day i met a big big big bear".split())
        assert self.test2.get_words() == set("i love love chickens because chickens are really cool chickens".split())
        assert self.test3.get_words() == set("gautham manigantan is a great human being with great characteristics".split())
        
    def test_get_path(self) -> None:
        assert self.doc1.get_path() == "doggos/doc1.txt"
        assert self.euro.get_path() == "small_wiki/Euro - Wikipedia.html"
        assert self.test1.get_path() == "test_files/test_1.txt"
        assert self.test2.get_path() == "test_files/test_2.txt"
        assert self.test3.get_path() == "test_files/test_3.txt"

    def test_repr(self) -> None:
        assert str(self.doc1) == "Document('doggos/doc1.txt')"
        assert str(self.euro) == "Document('small_wiki/Euro - Wikipedia.html')"
        assert str(self.test1) == "Document('test_files/test_1.txt')"
        assert str(self.test2) == "Document('test_files/test_2.txt')"
        assert str(self.test3) == "Document('test_files/test_3.txt')"
        
````

````python
%%ipytest -v

class SearchEngine:
    """
    Class that demonstrates a very simplified search engine over a specific file directory of documents
    which makes use of TF-IDF ranking to find a query.
    """

    def __init__(self, path: str, file_extension: str = ".txt") -> None:
        
        self._file_extension = file_extension
        self._path: str = path
        self._inverted: dict[str, set[str]] = {} # Dictionary that maps terms to a set of all the documents that they are in
        self._docs: dict[str, Document] = {} # Dictionary that maps terms to the documents that they are in

        # Going through every file in the folder
        for file in os.listdir(self._path):
            if (file.endswith(self._file_extension)): # If correct file type
                document = Document(os.path.join(self._path, file)) # Get the document
                self._docs[os.path.join(self._path, file)] = document

                # Every term in each document
                for term in document.get_words():

                    # Add them to the document set for that term if not already in there
                    if term not in self._inverted:
                        self._inverted[term] = set()
                    self._inverted[term].add(os.path.join(self._path, file))

    def _calculate_idf(self, term: str) -> float:
        """
        Takes a term and finds the frequency of it by computing the IDF
        """

        # If term is not in inverted list
        if clean(term) not in self._inverted.keys():
            return 0.0
        else:
            term_docs = len(self._inverted[clean(term)])
            num_docs = len(self._docs)
            return math.log(num_docs / term_docs) # total docs / num of docs with that term = idf value
    
    def __repr__(self) -> str:
        """
        Gives a string format for representing the search engine
        """

        return f"Search Engine(\'{self._path}\')"

    def search(self, search_query: str) -> list[str]:
        """
        Gives a list of document paths which are sorted by descending TF-IDF for the thing the user searched for in their query
        """

        search_words = search_query.split()
        doc_values: dict[str, float] = {}

        # Each word in the search query
        for term in search_words:
            idf = self._calculate_idf(term) # Calculate frequency for document occurrence across all documents
            for doc in self._inverted.get(clean(term), []):
                tf = self._docs[doc].term_frequency(term) # Calculate term frequency for each document
                doc_values[doc] = doc_values.get(doc, 0.0) + tf * idf # Compute the "score" for the document
        
        docs_sort = sorted(doc_values.items(), key = lambda item: item[1], reverse = True) # Sort by "Score" in reverse order
        return [doc for doc, value in docs_sort]

class TestSearchEngine:
    # Let mypy know these exist on instances:
    _doggos: SearchEngine
    _small_wiki: SearchEngine    
    _test_files: SearchEngine
    
    @classmethod
    def setup_class(cls) -> None:
        cls._doggos = SearchEngine("doggos")
        cls._small_wiki = SearchEngine("small_wiki", ".html")
        cls._test_files = SearchEngine("test_files")
    
    @staticmethod
    def os_path(path: str, filename: str) -> str:
        return os.path.join(path, filename)
    
    def test_search(self) -> None:
        assert self._doggos.search("love dogs") == [
            self.os_path("doggos","doc3.txt"), 
            self.os_path("doggos", "doc1.txt")]
        assert self._small_wiki.search("data")[:10] == [
            self.os_path("small_wiki", "Internet privacy - Wikipedia.html"),
            self.os_path("small_wiki", "Machine learning - Wikipedia.html"),
            self.os_path("small_wiki", "Bloomberg L.P. - Wikipedia.html"),
            self.os_path("small_wiki", "Waze - Wikipedia.html"),
            self.os_path("small_wiki", "Digital object identifier - Wikipedia.html"),
            self.os_path("small_wiki", "Chief financial officer - Wikipedia.html"),
            self.os_path("small_wiki", "UNCF - Wikipedia.html"),
            self.os_path("small_wiki", "Jackson 5 Christmas Album - Wikipedia.html"),
            self.os_path("small_wiki", "KING-FM - Wikipedia.html"),
            self.os_path("small_wiki", "The News-Times - Wikipedia.html"),
        ]
        assert self._test_files.search("i") == [
            self.os_path("test_files", "test_1.txt"),
            self.os_path("test_files", "test_2.txt")
        ]
````
